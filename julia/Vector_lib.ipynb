````julia
module Draw
using PyPlot
Fig,ax = missing,missing
vec= []
scale = 1
x_lim = [0,5]
y_lim = [0,5]
function draw_init end


function set_xyLabel end
function set_title end
function set_lim end
function set_scale end
function init_3d end
function drawing end

function clear_vec end
function del_vec end
function add_vec end 

mutable struct _Vector
    col::String
    vec::Vector
    function _Vector end
    
    _Vector(x::String,y::Vector) = new(x,y)
    _Vector() = _Vector("",[])
    _Vector(x::Vector) = _Vector("black",x)
    _Vector(y::String) = _Vector(y,[])
end

end
````

````julia

function Draw.set_lim(;x_lim = Draw.x_lim::Vector,y_lim = Draw.y_lim::Vector)
    Draw.ax.set_xlim(x_lim)
    Draw.ax.set_ylim(y_lim)
end

function Draw.set_lim(x_lim = Draw.x_lim::Vector,y_lim = Draw.ylim::Vector)
    Draw.ax.set_xlim(x_lim)
    Draw.ax.set_ylim(y_lim)
end


function Draw.set_xyLabel(X = "X",Y ="Y")
        Draw.plt.xlabel(X)
        Draw.plt.ylabel(Y)
end

function Draw.set_scale(scale)
    Draw.scale = scale
end

````

````julia

function Draw.del_vec(vec)
    for i in 1:length(Draw.vec)
        if i ==Draw.vec[i]
            delete!(Draw.vec,i)
            break
        end
    end
end

function Draw.clear_vec()
    Draw.vec = []
end
````

````julia

function Draw.drawing()
    #last = Draw.vec[end]
    #print(last[begin][1])
    #print(last[begin][1]," ",last[begin][2]," ",last[end],'\n')
    #Draw.ax.quiver(0,0,last[begin][1],last[begin][2],angles = "xy",scale_units = "xy",scale = Draw.scale,color = last[end])
        
    for i in 1:length(Draw.vec)
        last = Draw.vec[i]
        #print(last,'\n')
        Draw.ax.quiver(0,0,last[begin][1],last[begin][2],angles = "xy",scale_units = "xy",scale = Draw.scale,color = last[end])
        
    end     
end
````

````julia

function Draw.add_vec(vec)#vec 存储坐标与颜色
    
    flag = 0
    for i in vec
        if length(i)!=2 && length(i[1])!=2
            flag = 1
            break
        end
    end
        
    if flag == 1
        print("input error")
    end
        
    for i in vec
        push!(Draw.vec,i)
        #print(Draw.vec,"\n")
    end
end
Draw.add_vec() = Draw.add_vec([[[1,2],"black"]])
````

````julia

function Draw.draw_init()
    Draw.vec = []
    Draw.Fig,Draw.ax = Draw.plt.subplots()
    Draw.plt.grid(true)
    Draw.plt.xlabel("X")
    Draw.plt.ylabel("Y")
    Draw.x_lim = [0,5]
    Draw.y_lim = [0,5]
    Draw.ax.set_title("2d vectpr demo",color="blue")
    Draw.ax.set_xlim([-1,5])
    Draw.ax.set_ylim([-1,4])
       # print("vec init",vec)
end
````

````julia
function Draw.init_3d()
    Draw.Fig=Draw.plt.figure()
    Draw.ax = Draw.Fig.add_subplot(projection="3d")
    ax.set_title("3d vector")
    
    ax.grid(true)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    
    ax.set_xlim([0, 6])
    ax.set_ylim([0, 6])
    ax.set_zlim([0,6])
    
end
Draw.init_3d()
````

````julia
# 引入绘图库
using PyPlot
# 创建figure
fig = plt.figure()
# 在该figure中创建一个3维axes对象
ax = fig.add_subplot(projection="3d")
# 启用网格显示
ax.grid(true)
# 设置X轴标签
ax.set_xlabel("X")
# 设置Y轴标签
ax.set_ylabel("Y")
# 设置Z轴标签
ax.set_zlabel("Z")
# 定义三维向量
u = [1, 3, 5]
# 设置axes的标题及颜色
ax.set_title("3d vector demo", color="blue")
# 设置x轴取值范围为0~6
ax.set_xlim([0, 6])
# 设置y轴取值范围为0~6
ax.set_ylim([0, 6])
# 设置z轴取值范围为0~6
ax.set_zlim([0, 6])
# 绘制3维带箭头的线段来表示向量，起点为(0,0,0)，终点为(u[1],u[2],u[3])，颜色为红色
# arrow_length_ratio表示的末端箭头大小比率（0~1），它表示的箭头的长度和向量长度的比值，
# 可以试着调节该参数的大小看看效果
ax.quiver(0, 0, 0, u[1], u[2], u[3], color="red", arrow_length_ratio=0.1)
plt.show()
````
